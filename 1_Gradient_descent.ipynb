# 梯度縮減(Gradient Descent)
以y = x^2的範例，說明深度學習網路如何透過可微分之函數逼近目標最小值

```
import numpy as np
import matplotlib.pyplot as plt

# 定義目標函數：y = x^2
def func(x):
    return np.square(x)

# 目標函數的一階導數：dy/dx = 2 * x
def dfunc(x):
    return 1.5 * x

def GD(x_start, df, epochs, lr):
    """
    梯度下降法。給定起始點與目標函數的一階導函數，求在 epochs 次反覆運算中 x 的更新值
    :param x_start: x 的起始點
    :param df: 目標函數的一階導數
    :param epochs: 反覆運算的次數
    :param lr: 學習率
    :return: 每次反覆運算後 x 的位置，包含起始點，長度為 epochs + 1
    """
    # 初始化一個陣列來儲存每次更新的 x 值
    xs = np.zeros(epochs + 1)
    x = x_start
    xs[0] = x  # 將起始點放入 xs 陣列
    for i in range(epochs):
        dx = df(x)  # 計算 x 處的梯度
        v = - dx * lr  # 更新步長：負梯度方向乘以學習率
        x += v  # 更新 x 的值
        xs[i + 1] = x  # 將更新後的 x 值儲存在 xs 中
    return xs

# 主程式

# 設定參數
x_start = 10       # x 的初始值
epochs = 30       # 執行次數
lr = 0.6          # 學習率

# 執行梯度下降法
x = GD(x_start, dfunc, epochs, lr=lr)
print(x)

# 畫圖
color = 'r'  # 設定顏色

# 畫出目標函數的曲線
t = np.arange(-20.0, 10.0, 0.01)  # 設定 x 的範圍
plt.plot(t, func(t), c='b', label='y=x^2')  # 畫出 y = x^2 曲線

# 畫出每次更新後的 x 點
plt.plot(x, func(x), c=color, label=f'lr={lr}')
plt.scatter(x, func(x), c=color)  # 用點標出 x 的位置
plt.legend()  # 顯示圖例

plt.xlabel("x")  # 設定 x 軸標籤
plt.ylabel("y")  # 設定 y 軸標籤
plt.title("Helloteacher")  # 設定圖表標題
plt.show()  # 顯示圖表
```
